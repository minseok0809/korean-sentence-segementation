#### AIHUB Preprocessing Flowchart

```python
schemdraw.config( font='NanumGothic')

with schemdraw.Drawing() as d:
    #make_json_txt_file_path_list
    d+= Start(w = 12).label("원천 데이터 JSON 파일과 추출 TXT 파일 경로 리스트 생성")
    d+= Arrow().down(d.unit/2)

    #write_jsontext_to_txt_file_with_batch_list
    d+= Box(w = 12,  h=2.5).label("JSON에서 추출한 텍스트를 Batch Size로 TXT 파일에 나누어 저장\n\n TXT 파일 개수 Count에 따라 Batch Size 결정")
    d+= Arrow().down(d.unit/2)


    #make_pro_post_txt_file_path_list
    d+= Data(w = 10).label("전처리 이전과 이후 TXT 파일 경로 리스트 생성")
    d+= Arrow().down(d.unit/2)
    
    #preprocessing_corpus_txt
    d+= Box(w = 11.5, h=3).label("TXT 파일 1개씩 전처리 작업 수행\n\n- 작업 진행 상황 확인: 현재까지 전처리한 TXT 파일 개수 Print\n- 병렬 처리: 속도 향상")
    d+= Arrow().down(d.unit/2)
    
    #preprocessing_corpus_txt
    d+= Ellipse(w = 12).label("전처리 이후 TXT 파일 병합 및 중복 문장 제거")

    d.save("aihub flowchart.jpeg", dpi = 900)
```

```python
schemdraw.config( font='NanumGothic')

with schemdraw.Drawing() as d:
    d+= Start(w = 13).label("원천 데이터 JSON 파일 1개에서 텍스트를 추출하여 리스트로 저장")
    d+= Arrow().down(d.unit/2)

    #Check if string and reverse_string are same
    d+= (decision := Decision(w = 12, h= 3,
                        S = "True",
                        E = "False").label("len(list) >= Batch Size"))

    #If True
    d+= Arrow().length(d.unit/2)
    d+= (true := Box(w = 8).label("전체 리스트를 Batch Size에 따라 분할"))
    d+= Arrow().length(d.unit/2)

    #write_jsontext_to_txt_file_with_batch_list
    d+= Box(w = 8).label("각 분할된 리스트를 TXT 파일 여러 개에 저장")
    d+= Arrow().down(d.unit/2)
    
        #End program
    d+= (end := Ellipse().label("End"))
    

    #If False. Start the arrow from East of decision box
    d+= Arrow().right(d.unit).at(decision.E)

    #false is referring to the box where string is not a palindrome.
    d+= (false := (decision := Decision(w = 12, h= 3,
                        N = "True").label("len(list) is very small")))

    #If True
    d+= Arrow().up(d.unit/2.3).at(false.N)
    d+= (false := (true := Box(w = 12, h= 1.5).label("len(list)가 Batch Size 이상일 때 까지 JSON 파일에서 리스트 병합")))



    #Extend the arrow to reach the end of the program
    d+= Arrow().left(d.unit/1.2).at(false.W)
    d.save("aihub flowchart3.jpeg", dpi = 300)
```

```python
schemdraw.config( font='NanumGothic')

with schemdraw.Drawing() as d:
    d+= Start(w = 13).label("원천 데이터 JSON 파일 1개에서 텍스트를 추출하여 리스트로 저장")
    d+= Arrow().down(d.unit/2)

    #Check if string and reverse_string are same
    d+= (decision := Decision(w = 12, h= 3,
                        S = "True",
                        E = "False").label("len(list) >= Batch Size"))

    #If True
    d+= Arrow().length(d.unit/2)
    d+= (true := Box(w = 8).label("전체 리스트를 Batch Size에 따라 분할"))
    d+= Arrow().length(d.unit/2)

    #write_jsontext_to_txt_file_with_batch_list
    d+= Box(w = 8).label("각 분할된 리스트를 TXT 파일 여러 개에 저장")
    d+= Arrow().down(d.unit/2)
    
        #End program
    d+= (end := Ellipse().label("End"))
    

    #If False. Start the arrow from East of decision box
    d+= Arrow().right(d.unit).at(decision.E)

    #false is referring to the box where string is not a palindrome.
    d+= (false := Box(w = 8).label("전체 리스트를 TXT 파일 1개에 저장"))

    #Add a downward arrow from the South of false box 
    d+= Arrow().down(d.unit*2.5).at(false.S)


    #Extend the arrow to reach the end of the program
    d+= Arrow().left(d.unit*3)
    d.save("aihub flowchart4.jpeg", dpi = 300)
```
